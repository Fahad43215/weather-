```python
import pandas as pd
datas=pd.read_csv('Comilla.csv')
datas.info()
```

```python



missing_data =datas.isnull().sum()
print(missing_data)
```

```python
# Ensure Date-Time is in datetime format
datas['Date-Time'] = pd.to_datetime(datas['Date-Time'])

# Filter for the month you need (e.g., January)
month_data = datas[datas['Date-Time'].dt.month == 1]

print(month_data)

```

```python
# Create a DataFrame for missing data table
missing_data_table = pd.DataFrame({
    "Column Name": month_data.columns,
    "Missing Count": month_data.isnull().sum().values,
    "Missing Percentage (%)": (month_data.isnull().mean() * 100).round(2).values
})

print(missing_data_table)
```

```python
# Ensure Date-Time is in datetime format
datas['Date-Time'] = pd.to_datetime(datas['Date-Time'])

# Group by month and calculate missing values
monthly_missing_data = datas.groupby(datas['Date-Time'].dt.month).apply(lambda x: pd.DataFrame({
    "Column Name": x.columns,
    "Missing Count": x.isnull().sum().values,
    "Missing Percentage (%)": (x.isnull().mean() * 100).round(2).values
}))

# Display the results
for month in range(1, 13):
    print(f"\nMissing Data for Month {month}:\n")
    print(monthly_missing_data.loc[month])
```

```python
# Ensure Date-Time is in datetime format
datas['Date-Time'] = pd.to_datetime(datas['Date-Time'])

# Define month names
month_names = {1: "January", 2: "February", 3: "March", 4: "April",
               5: "May", 6: "June", 7: "July", 8: "August",
               9: "September", 10: "October", 11: "November", 12: "December"}


monthly_summary = datas.groupby(datas['Date-Time'].dt.month).apply(lambda x: pd.Series({
    "Month": month_names[x['Date-Time'].dt.month.iloc[0]],  # Convert month number to name
    "Total Data Points": x.size,  # Total number of values in the month
    "Total Missing Count": x.isnull().sum().sum(),  # Sum of all missing values
    "Total Missing Percentage (%)": round((x.isnull().sum().sum() / x.size) * 100, 2)  # Percentage of missing data
})).reset_index(drop=True)

# Display the summary table
print(monthly_summary)
```
